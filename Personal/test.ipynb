### ä»€ä¹ˆæ˜¯æ¥å£(interface)

å°†å¯¹è±¡çš„ä¸€éƒ¨åˆ†åŠŸèƒ½ç‹¬ç«‹æŠ½è±¡å‡ºæ¥çš„äº§ç‰©ã€‚å…·æœ‰ç›¸åŒæ–¹æ³•çš„ç±»å‹ï¼Œæ„å‘³ç€å¯ä»¥ä¸ç®¡å®ƒä»¬çš„ç±»å‹å¦‚ä½•ï¼ŒæŒ‰ç…§åŒæ ·çš„æ–¹å¼è°ƒç”¨è¿™ä¸ªæ–¹æ³•ã€‚é‚£ä¹ˆéœ€è¦è¿™ä¸ªæ–¹æ³•çš„åœ°æ–¹ï¼Œå¯ä»¥ç”¨æ¥å£ä»£æ›¿ã€‚

> å®é™…ä¸Š, Python ä¸å­˜åœ¨æ¥å£çš„æ¦‚å¿µï¼Œå› ä¸ºPython å±äºåŠ¨æ€è¯­è¨€ï¼Œå¯¹ç±»å‹åˆ¤å®šå¹¶æ²¡æœ‰é™æ€è¯­è¨€é‚£æ ·ä¸¥æ ¼

```python
from typing import Protocol


class Walkable(Protocol):
    """å…·æœ‰Walk æ–¹æ³•çš„æ¥å£"""
    def walk(self):
        ...

class Pig(Walkable):
    """ç»§æ‰¿äº†Walkable çš„å­ç±»"""
    def walk(self):
        print("pig walking.")

class Rabbit:
    """å®é™…ä¸Šï¼Œä¸éœ€è¦æ˜¾ç¤ºç»§æ‰¿ä¹Ÿè¡Œ"""
    def walk(self):
        print("rabbit walking.")
```

```python
def func_need_some_walkable(instance: Walkable):
    """éœ€è¦ä¸€ä¸ªæ‹¥æœ‰ walk æ–¹æ³•çš„å¯¹è±¡"""
    instance.walk()
```

```python
pig = Pig()
# Dog æ‹¥æœ‰ walk() æ–¹æ³•
func_need_some_walkable(pig)
```

```python
rab = Rabbit()
# Cat ä¹Ÿå…·æœ‰ walk() æ–¹æ³•
func_need_some_walkable(rab)
```
